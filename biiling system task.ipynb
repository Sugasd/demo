```python
from tkinter import *
import time
from tkinter import messagebox
import os
from random import randint
color='#87918c'
color1='#757d79'
color3='#bf5454'      #exit button
color4='#87918c'
color5='#87918c'
color6='#87918c'
x = randint(10, 400)
path = 'E:/SD'  # path to the folder
root = Tk()
root.geometry("1150x750")
root.title("billing system")
Tops = Frame(root, width=1350, height=50, bd=8, bg=color, relief="raise")
Tops.pack(side=TOP)
Bottoms = Frame(root, width=1350, height=50, bd=8, bg=color, relief="raise")
Bottoms.pack(side=BOTTOM)
f1 = Frame(root, width=900, height=650, bd=8, bg=color, relief="raise")
f1.pack(side=LEFT)
f1a = Frame(f1, width=900, height=330, bd=8, bg=color, relief="raise")
f1a.pack(side=TOP)
f2a = Frame(f1, width=900, height=320, bd=8, bg=color, relief="raise")
f2a.pack(side=BOTTOM)
f1aa = Frame(f1a, width=400, height=430, bd=8, bg=color4, relief="raise")
f1aa.pack(side=LEFT)
f1ab = Frame(f1a, width=400, height=430, bd=8, bg=color5, relief="raise")
f1ab.pack(side=RIGHT)
f2aa = Frame(f2a, width=450, height=330, bd=8, bg=color6, relief="raise")
f2aa.pack(side=LEFT)
f2ab = Frame(f2a, width=450, height=330, bd=8, bg=color4, relief="raise")
f2ab.pack(side=LEFT)
lblInfo = Label(Tops, font=('arial', 25, 'bold'), text="SHERIFF BHAI BIRYANI", bg="pale goldenrod",bd=15,padx=12, anchor='w')
lblInfo.grid(row=0, column=0)


#Variables Declarations


PaymentRef = StringVar()
emailID = StringVar()
mushroom_biryani = StringVar()
chicken_biryani = StringVar()
mutton_biryani = StringVar()
prawn_biryani = StringVar()
costmsb = StringVar()
costchb = StringVar()
costmtb = StringVar()
costpnb = StringVar()
dateRef = StringVar()
subTotal = StringVar()
discount = StringVar()
totalPrice = StringVar()
text_Input = StringVar()
dateRef.set(time.strftime("%d/%m/%y"))
operator = ""
discount.set(0)
mushroom_biryani.set(0)
chicken_biryani.set(0)
mutton_biryani.set(0)
prawn_biryani.set(0)
subTotal.set(0)
totalPrice.set(0)
costmsb.set(110)
costchb.set(140)
costmtb.set(220)
costpnb.set(170)

# Functions

def Price():
    msbprice = int(costmsb.get())
    chbprice = int(costchb.get())
    mtbprice = int(costmtb.get())
    pnbprice = int(costpnb.get())
    msbno = int(mushroom_biryani.get())
    chbno = int(chicken_biryani.get())
    mtbno = int(mutton_biryani.get())
    pnbno = int(prawn_biryani.get())
    tempdiscount = int(discount.get())
    subPrice = (msbprice * msbno + chbprice * chbno + mtbprice * mtbno + pnbprice * pnbno)
    totalCost = str('%d' % subPrice)
    totalCostwithdiscount = str('%d' % (subPrice - (subPrice * tempdiscount) / 100))
    subTotal.set(totalCost)
    totalPrice.set(totalCostwithdiscount)
def Exit():
    qexit = messagebox.askyesno("Billing System", "Do you want to exit?")
    if qexit > 0:
        root.destroy()
        return
def reset():
    global x
    x = x + 1
    PaymentRef.set("")
    mushroom_biryani.set(0)
    chicken_biryani.set(0)
    mutton_biryani.set(0)
    prawn_biryani.set(0)
    subTotal.set(0)
    totalPrice.set(0)
def refNo():
    global x
    y = str(x)
    randomRef = str(y)
    PaymentRef.set("BILL" + randomRef)
def create_bill():
    global x
    refno = str(x)
    file = refno + ".txt"
    global path
    with open(os.path.join(path, file), "w") as file1:
        toFile = output()
        file1.write(toFile)
    qmsg = messagebox.showinfo("Information", "Bill Generated")

def output():
    global x
    refno = str(x)
    list0 = "\t\t\t\t\tReference.no :" + refno+"\n"
    list02= "\t\t SHERIFF BHAI BRIYANI"
    list1 = "\n\n" + "Item\t\t\tQuantity\t\tCost\n"
    list7 = "____\t\t\t_______\t\t        ____\n\n"
    list2 = "MUSHROOM BIRYANI\t" + mushroom_biryani.get() + "\t\t\t" + str(int(mushroom_biryani.get()) * int(costmsb.get())) + "\n"
    list3 = "CHICKEN BIRYANI\t\t" + chicken_biryani.get() + "\t\t\t" + str(int(chicken_biryani.get()) * int(costchb.get())) + "\n"
    list4 = "MUTTON BIRYANI\t\t" + mutton_biryani.get() + "\t\t\t" + str(int(mutton_biryani.get()) * int(costmtb.get())) + "\n"
    list5 = "PRAWN BIRYANI\t\t" + prawn_biryani.get() + "\t\t\t" + str(int(prawn_biryani.get()) * int(costpnb.get())) + "\n\n"
    list6 = "\t\t\t   " + "Total     = Rs " + subTotal.get() + "/-" + "\n"
    list8 = "\t\t\t   " + "Discount  = Rs " + str(int(totalPrice.get()) - int(subTotal.get())) + "/-" + "\n"
    list9 = "\t\t\t   " + "GrandTotal= Rs " + totalPrice.get()[:] + "/-" + "\n"
    String = list0 + list02 + list1 + list7 + list2 + list3 + list4 + list5 + list6 + list8 + list9
    return String

# Order Information

lblRef = Label(f1aa, font=('arial', 16, 'bold'), fg="red", text="Reference No", bd=16, bg=color4, justify='left')
lblRef.grid(row=0, column=0)
txtRef = Entry(f1aa, font=('arial', 16, 'bold'), textvariable=PaymentRef, bd=10, insertwidth=2, justify='left')
txtRef.grid(row=0, column=1)
# --------------

lblCb = Label(f1aa, font=('arial', 16, 'bold'), text="MUSHROOM BIRYANI", bd=16, bg=color4, justify='left')
lblCb.grid(row=1, column=0)
txtCb = Entry(f1aa, font=('arial', 16, 'bold'), textvariable=mushroom_biryani, bd=10, insertwidth=2, justify='left')
txtCb.grid(row=1, column=1)
# --------------

lblBb = Label(f1aa, font=('arial', 16, 'bold'), text="CHICKEN BIRYANI", bd=16, bg=color4, justify='left')
lblBb.grid(row=2, column=0)
txtBb = Entry(f1aa, font=('arial', 16, 'bold'), textvariable=chicken_biryani, bd=10, insertwidth=2, justify='left')
txtBb.grid(row=2, column=1)
# --------------

lblFf = Label(f1aa, font=('arial', 16, 'bold'), text="MUTTON BIRYANI", bd=16, bg=color4, justify='left')
lblFf.grid(row=3, column=0)
txtFf = Entry(f1aa, font=('arial', 16, 'bold'), textvariable=mutton_biryani, bd=10, insertwidth=2, justify='left')
txtFf.grid(row=3, column=1)
# --------------

lblSd = Label(f1aa, font=('arial', 16, 'bold'), text="PRAWN BIRYANI", bd=16, bg=color4, justify='left')
lblSd.grid(row=4, column=0)
txtSd = Entry(f1aa, font=('arial', 16, 'bold'), textvariable=prawn_biryani, bd=10, insertwidth=2, justify='left')
txtSd.grid(row=4, column=1)


# Price  Information

lbldate = Label(f1ab, font=('arial', 16, 'bold'), text="Date", bd=16, bg=color5, justify='left')
lbldate.grid(row=0, column=0)
txtdate = Entry(f1ab, font=('arial', 16, 'bold'), textvariable=dateRef, bd=10, insertwidth=2, justify='left')
txtdate.grid(row=0, column=1)
# --------------

lblCcb = Label(f1ab, font=('arial', 16, 'bold'), text="Price of Mushroom Biryani", bd=16, bg=color5, justify='left')
lblCcb.grid(row=1, column=0)
txtCcb = Entry(f1ab, font=('arial', 16, 'bold'), textvariable=costmsb, bd=10, insertwidth=2, justify='left')
txtCcb.grid(row=1, column=1)
# --------------

lblCbb = Label(f1ab, font=('arial', 16, 'bold'), text="Price of Chicken Biryani", bd=16, bg=color5, justify='left')
lblCbb.grid(row=2, column=0)
txtCbb = Entry(f1ab, font=('arial', 16, 'bold'), textvariable=costchb, bd=10, insertwidth=2, justify='left')
txtCbb.grid(row=2, column=1)
# --------------

lblCff = Label(f1ab, font=('arial', 16, 'bold'), text="Price of Mutton Biryani", bd=16, bg=color5,
               justify='left')
lblCff.grid(row=3, column=0)
txtCff = Entry(f1ab, font=('arial', 16, 'bold'), textvariable=costmtb, bd=10, insertwidth=2,
               justify='left')
txtCff.grid(row=3, column=1)
# --------------

lblCsd = Label(f1ab, font=('arial', 16, 'bold'), text="Price of Prawn Biryani", bd=16, bg=color5, justify='left')
lblCsd.grid(row=4, column=0)
txtCsd = Entry(f1ab, font=('arial', 16, 'bold'), textvariable=costpnb, bd=10, insertwidth=2, justify='left')
txtCsd.grid(row=4, column=1)


# Total Amount Information
lblPrice = Label(f2aa, font=('arial', 16, 'bold'), text="Price", bd=16, bg=color6, justify='left')
lblPrice.grid(row=0, column=0)
txtPrice = Entry(f2aa, font=('arial', 16, 'bold'), textvariable=subTotal, bd=10, insertwidth=2, justify='left')
txtPrice.grid(row=0, column=1)
# --------------

lbldiscount = Label(f2aa, font=('arial', 16, 'bold'), text="Discount", bd=16, bg=color6, justify='left')
lbldiscount.grid(row=1, column=0)
txtdiscount = Entry(f2aa, font=('arial', 16, 'bold'), textvariable=discount, bd=10, insertwidth=2, justify='left')
txtdiscount.grid(row=1, column=1)
# --------------

lblTp = Label(f2aa, font=('arial', 16, 'bold'), text="Total Price", bd=16, bg=color6, justify='left')
lblTp.grid(row=2, column=0)
txtTp = Entry(f2aa, font=('arial', 16, 'bold'), textvariable=totalPrice, bd=10, insertwidth=2, justify='left')
txtTp.grid(row=2, column=1)


# Buttons
btnTotal = Button(f2ab, padx=16, pady=16, bd=8, fg="black", font=('arial', 16, 'bold'), width=15,
                  text="Total Price", bg=color, command=Price).grid(row=1, column=0)
btnRefer = Button(f2ab, padx=16, pady=16, bd=8, fg="black", font=('arial', 16, 'bold'), width=15,
                  text="Sales Reference", bg=color, command=refNo).grid(row=0, column=0)
btnCrtb = Button(f2ab, padx=16, pady=16, bd=8, fg="black", font=('arial', 16, 'bold'), width=15,
                 text="Generate bill", bg=color1, command=create_bill).grid(row=1, column=1)
btnReset = Button(f2ab, padx=16, pady=16, bd=8, fg="black", font=('arial', 16, 'bold'), width=15,
                  text="Reset", bg=color, command=reset).grid(row=0, column=1)
btnExit = Button(f2ab, padx=16, pady=16, bd=8, fg="black", font=('arial', 16, 'bold'), width=10,
                 text="Exit", bg=color3, command=Exit).grid(row=0, column=2)
root.mainloop()
```
